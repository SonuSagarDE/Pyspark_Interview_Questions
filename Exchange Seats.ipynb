```
data=[(1,'Alice'),(2,'Bob'),(3,'Sam'),(4,'Jay'),(5,'Jazz')]
schema=['id','student']
df=spark.createDataFrame(schema=schema,data=data)
```

```
display(df)
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import lead,lag,col,when,coalesce
windowspec=Window.orderBy('id')

df1=df.withColumn('prev',lag("student").over(windowspec))\
        .withColumn('Nxt',lead("student").over(windowspec))
df1.show()        
```

```
df2=df1.withColumn('Exchanged_value',when(col("id")%2==0,coalesce(col("prev"),col("student")) )\
                                     .when(col("id")%2==1,coalesce(col("Nxt"),col("student")) )\
                                      .otherwise(col("student")))
                             
```

```
df2.select("id","Exchanged_value").show()
```
